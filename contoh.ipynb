```python
#contoh latihan tuple untuk mencari item terbanyak

import string
membuka = open("romeo-full.txt")
hitung = dict()
for baris in membuka:
    baris = baris.translate(str.maketrans("","",string.punctuation))
    baris = baris.lower()
    kata = baris.split()
    for i in kata:
        if i not in hitung:
            hitung[i] = 1
        else:
            hitung[i] += 1

Listnya = list()

for key,val in list(hitung.items()):
    Listnya.append((val,key))
Listnya.sort(reverse=True)

for key in Listnya[:]:
    print(key)
```
